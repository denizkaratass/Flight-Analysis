# 4. Departure & Arrival Time Analysis
Visualizing the preferred times for take-off and landing. We compare different times of the day (Morning, Night, Afternoon) to see flight frequency patterns.

```python
df.head(2)  
 
```

```python
x=df["departure_time"].value_counts()
x
type(x)
```

```python
df["arrival_time"].value_counts()
```

```python
plt.figure(figsize= (16,4))

plt.subplot(1,2,1)
#It divides the figure into 1 row and 2 columns, and selects the 1st plot (left side) to draw.
plt.bar(df["departure_time"].value_counts().index ,df["departure_time"].value_counts().values, color= ["lightgrey","lightpink"]);
plt.title("Departure Time")
plt.xlabel(" D.Time")
plt.ylabel(" D.Frequency")

plt.subplot(1,2,2) #select 2nd plot to drow
plt.bar(df["arrival_time"].value_counts().index ,df["arrival_time"].value_counts().values, color= ["lightgrey","lightblue"]);
plt.title("Arrival Time")
plt.xlabel(" A.Time")
plt.ylabel(" A.Frequency")

plt.show()
```

# 6. Price Analysis by Airline
Calculating the average ticket price for each airline. We also use `catplot` to visualize the price distribution, differentiating between **Economy** and **Business** class.

```python
df.head(2)
```

```python
df.groupby("airline")["price"].mean()
```

```python
plt.figure(figsize= (16,4))
plt.bar(df.groupby("airline")["price"].mean().sort_values(ascending=True).index ,df.groupby("airline")["price"].mean().sort_values(ascending=True).values ,color=["lightgreen"]);
plt.title("Average Ticket Price by Airlines")
plt.ylabel("average price")
plt.xlabel(" name of airlines")
plt.show()
```

```python
sns.catplot( x="airline",y="price", kind="bar", palette="rocket", data=df)
plt.show()
```

```python
sns.catplot( x="airline",y="price", kind="bar", palette="rocket", data=df ,hue="class")
plt.show()
```
